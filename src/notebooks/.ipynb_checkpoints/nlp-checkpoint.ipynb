# Preproc

```python
import functools
import re
```

```python
url_pattern = re.compile(r'https?://\S+|www\.\S+')
mention_pattern = re.compile(r'@\w+')
nonword_pattern = re.compile(r'[^а-яА-ЯёЁa-zA-Z\s.,]+')
wikiref_pattern = re.compile(r'(\[\d\])')

def remove_urls(text):
    return url_pattern.sub(r' ', text)

def remove_mentions(text):
    return mention_pattern.sub(r' ', text)

def remove_nonwords(text):
    return nonword_pattern.sub(r' ', text)

def remove_wikiref(text):
    return wikiref_pattern.sub(r' ', text)

def remove_extra_spaces(text):
    return re.sub(' +', ' ', text).strip()
    
def preprocess_text(text, filters):
    return functools.reduce(lambda prew, f: f(prew), filters, text)

# TODO: OPECHATKI
```

## REBEL

```python
def extract_triplets_typed(text):
    triplets = []
    relation = ''
    text = text.strip()
    current = 'x'
    subject, relation, object_, object_type, subject_type = '','','','',''

    for token in text.replace("<s>", "").replace("<pad>", "").replace("</s>", "").replace("tp_XX", "").replace("__ru__", "").split():
        if token == "<triplet>" or token == "<relation>":
            current = 't'
            if relation != '':
                triplets.append({'head': subject.strip(), 'head_type': subject_type, 'type': relation.strip(),'tail': object_.strip(), 'tail_type': object_type})
                relation = ''
            subject = ''
        elif token.startswith("<") and token.endswith(">"):
            if current == 't' or current == 'o':
                current = 's'
                if relation != '':
                    triplets.append({'head': subject.strip(), 'head_type': subject_type, 'type': relation.strip(),'tail': object_.strip(), 'tail_type': object_type})
                object_ = ''
                subject_type = token[1:-1]
            else:
                current = 'o'
                object_type = token[1:-1]
                relation = ''
        else:
            if current == 't':
                subject += ' ' + token
            elif current == 's':
                object_ += ' ' + token
            elif current == 'o':
                relation += ' ' + token
    if subject != '' and relation != '' and object_ != '' and object_type != '' and subject_type != '':
        triplets.append({'head': subject.strip(), 'head_type': subject_type, 'type': relation.strip(),'tail': object_.strip(), 'tail_type': object_type})
    return triplets

# Load model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("Babelscape/mrebel-base", src_lang="ru", tgt_lang="ru")

# Here we set English ("en") as source language. To change the source language swap the first token of the input for your desired language or change to supported language. 
model = AutoModelForSeq2SeqLM.from_pretrained("Babelscape/mrebel-base")
gen_kwargs = {
    "max_length": 256,
    "length_penalty": 0,
    "num_beams": 3,
    "num_return_sequences": 3,
    "forced_bos_token_id": None,
}

# Tokenizer text
model_inputs = tokenizer(text, max_length=256, padding=True, truncation=True, return_tensors = 'pt')

# Generate
generated_tokens = model.generate(
    model_inputs["input_ids"].to(model.device),
    attention_mask=model_inputs["attention_mask"].to(model.device),
    **gen_kwargs,
)

# Extract text
decoded_preds = tokenizer.batch_decode(generated_tokens, skip_special_tokens=False)

# Extract triplets
triplets = list()
for idx, sentence in enumerate(decoded_preds):
    print(f'Prediction triplets sentence {idx}')
    triplet = extract_triplets_typed(sentence)
    print(triplet)
    triplets = triplets + triplet
```

```python
ents = set()
for t in triplets:
    ents.add(t['head'])
    ents.add(t['tail'])
ents
```

#### rubert-tiny2

```python
tokenizer = transformers.AutoTokenizer.from_pretrained("cointegrated/rubert-tiny")
model = transformers.AutoModel.from_pretrained("cointegrated/rubert-tiny")
```

```python
def embed_bert_cls(text, model, tokenizer):
    t = tokenizer(text, padding=True, truncation=True, return_tensors='pt')
    with torch.no_grad():
        model_output = model(**{k: v.to(model.device) for k, v in t.items()})
    embeddings = model_output.last_hidden_state[:, 0, :]
    embeddings = torch.nn.functional.normalize(embeddings)
    return embeddings[0].cpu().numpy()
```

```python
df['cointegrated/rubert-tiny'] = df['span'].apply(lambda t: embed_bert_cls(t, model, tokenizer))
qdf['cointegrated/rubert-tiny'] = qdf['questions'].apply(lambda t: embed_bert_cls(t, model, tokenizer))
```

```python
def is_sim_col(col_name, model_name, question):
    return (model_name in col_name) and (question in col_name)
```

```python
# {1,2, 3} & {1,5,6}
```

```python
metrics = ['ndcg_score', 'Precision@K', 'Recall@K', 'F1@K']
```

```python
def precision_at_k(indexes, ideal, k=3):
    indexes = list(indexes)
    got = set(indexes[:k])
    want = set(ideal)
    rel_num = len(got&want)
    return rel_num / k

precision_at_k(resdf['index'], [0, 1, 2]), precision_at_k(resdf['index'], [0, -1, 2])
```

```python
def recall_at_k(indexes, ideal, k=3, anyval=-1):
    indexes = list(indexes)
    got = set(indexes[:k])
    want = set([i for i in ideal if i != anyval])
    rel_num = len(got&want)
    return rel_num / len(want)

recall_at_k(resdf['index'], [0, -1, 2]), recall_at_k(resdf['index'], [0, -1, 4])
```

```python
def at_k_scores(indexes, ideal, k=3, anyval=-1, beta=1):
    precision = precision_at_k(indexes, ideal, k)
    recall = recall_at_k(indexes, ideal, k, anyval)
    if precision + recall == 0:
        return [0, 0, 0]
    f = ((1+beta**2) * precision * recall) / (beta**2*precision + recall)
    return [f, precision, recall]

at_k_scores(resdf['index'], [0, -1, 2])
```

```python
def test(
    df, 
    model_name, 
    question, 
    
    ideal,
    true_relevance, 
    relevance,
    
    k=3,
    anyval=-1,
    beta=1,
    
    limit=10,
):
    print(f'{model_name} | {question}\n')
    sims_keys = [k for k in df.keys() if is_sim_col(k, model_name, question)]
    keys = ['index', 'span'] + sims_keys
    df[keys]
    t = df[keys].sort_values(by=sims_keys[0], ascending=False).iloc[:limit]
    for i in range(5):
        r = t[keys].iloc[i]
        text = r['span']
        sim = r[sims_keys[0]]
        print(f'{r["index"]}\n{text.strip()}\n{sim}\n\n')

    indexes = list(t['index'])
    print(f'{ideal=} {indexes=}')
    ndcg_score_res = ndcg_score(true_relevance, relevance, k=5)
    at_k_scores_res = at_k_scores(indexes, ideal, k, anyval, beta)
    print(metrics)
    return [ndcg_score_res] + at_k_scores_res
```
